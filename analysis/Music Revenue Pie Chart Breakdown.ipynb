```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
music_data = pd.read_csv("data_input/Revenue_Chart_Full_Data_data.csv")
music_data
```

```python
df_1985 = music_data.loc[music_data["Year of Year Date"].isin([1985])]
df_1991 = music_data.loc[music_data["Year of Year Date"].isin([1991])]
df_1999 = music_data.loc[music_data["Year of Year Date"].isin([1999])]
df_2005 = music_data.loc[music_data["Year of Year Date"].isin([2005])]
df_2014 = music_data.loc[music_data["Year of Year Date"].isin([2014])]
df_2022 = music_data.loc[music_data["Year of Year Date"].isin([2022])]
```

```python
df_1985
```

```python
d_1985 = {"chart_labels": ["CD", "Cassette", "LP/EP"]}
```

```python
labels = ["CD", "Cassette", "LP/EP",]

sizes = [389.5, 2411.5, 1561.5]

colors = ["olive", "darkolivegreen", "darkkhaki"]

explode = (.1, 0, 0)
```

```python
plt.pie(sizes,labels=d_1985["chart_labels"],startangle=110,explode=explode,colors=colors,shadow=False, autopct="%1.2f%%")
plt.axis("square")
plt.savefig("graph_output/1985_piegraph.png", bbox_inches="tight")
```
