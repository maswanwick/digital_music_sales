```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
gym_membership_pd = pd.read_csv('output/gym_membership_by_year.csv')
digital_revenue_pd = pd.read_csv('output/physical_vs_digital_revenue.csv')

gym_revenue_pd = pd.merge(gym_membership_pd, digital_revenue_pd, left_on='year', right_on='Year')

gym_revenue_pd
```

```python
gym_revenue_pd.plot(kind="scatter", x="membership_count", y="Total Digital Revenue")
plt.xlabel("U.S. Gym Memberships (in millions)")
plt.ylabel("Digital Music Revenue (in millions)")

(slope, intercept, rvalue, pvalue, stderr) = linregress(gym_revenue_pd['membership_count'], gym_revenue_pd['Total Digital Revenue'])
regress_values = gym_revenue_pd['membership_count'] * slope + intercept
plt.plot(gym_revenue_pd['membership_count'], regress_values, "r-")

plt.show()

print(rvalue**2)
```
